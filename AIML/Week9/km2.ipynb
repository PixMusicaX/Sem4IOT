```python
from sklearn.datasets import load_iris

# Loading the iris dataset from sklearn.datasets
iris = load_iris()

# Displaying the description of the iris dataset
print(iris.DESCR)
```

```python
import pandas as pd

# Keeping the data in a dataframe
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Selecting the features sepal_length, sepal_width, petal_length and petal_width for clustering
X = df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']]
```

```python
import matplotlib.pyplot as plt

# Plotting the selected data as a scatter plot using matplotlib
plt.scatter(X['sepal length (cm)'], X['sepal width (cm)'], c='blue')
plt.scatter(X['petal length (cm)'], X['petal width (cm)'], c='red')
plt.xlabel('Length (cm)')
plt.ylabel('Width (cm)')
plt.show()
```

```python
from sklearn.cluster import KMeans

# Using the kMeans clustering algorithm for finding 3 clusters from the generated data
kmeans = KMeans(n_clusters=3)

# Fitting the model to our data
kmeans.fit(X)

# Determining and recording the cluster predictions and cluster centres
y_kmeans = kmeans.predict(X)
centers = kmeans.cluster_centers_

# Plotting the clusters with respective cluster centres (use different colors for each cluster and black for the cluster centres)
plt.scatter(X['sepal length (cm)'], X['sepal width (cm)'], c=y_kmeans)
plt.scatter(X['petal length (cm)'], X['petal width (cm)'], c=y_kmeans)
plt.scatter(centers[:,0], centers[:,1], c='black', s=200, alpha=0.5)
plt.xlabel('Length (cm)')
plt.ylabel('Width (cm)')
plt.show()
```
