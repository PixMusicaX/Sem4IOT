```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generating data-set of 200 samples with 3 centers, 2 features, cluster standard-deviation of 1.5 and random state = 30
X, y = make_blobs(n_samples=200, centers=3, n_features=2, cluster_std=1.5, random_state=30)

# Plotting the generated data as a scatter plot using matplotlib
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
from sklearn.cluster import KMeans

# Using the kMeans clustering algorithm for finding 3 clusters from the generated data
kmeans = KMeans(n_clusters=3)

# Fitting the model to our data
kmeans.fit(X)

# Determining and recording the cluster predictions and cluster centres
y_kmeans = kmeans.predict(X)
centers = kmeans.cluster_centers_

# Plotting the clusters with respective cluster centres (use different colors for each cluster and black for the cluster centres)
plt.scatter(X[:,0], X[:,1], c=y_kmeans)
plt.scatter(centers[:,0], centers[:,1], c='black', s=200, alpha=0.5)
plt.show()
```
